### *Ejemplo de ecuaciones*

El balance energético global, suponiendo un sistema adiabático en régimen estacionario, se puede plantear de la siguiente manera,

\begin{equation}
0= \dot{Q}+\dot{W} + \sum(\dot{m}_i h_i)_e-\sum(\dot{m}_i h_i)_s = 0 + 0 + \dot{m}_{a1}h_{a1} +\dot{m}_{w1}h_{w1}-\dot{m}_{a2}h_{a2}-\dot{m}_{w2}h_{w2}.
\end{equation}

Mientras que los balances másicos resultan:

\begin{equation}
\dot{m}_{a1}=\dot{m}_{a2}
\end{equation}

\begin{equation}
\dot{m}_{a1}w_1 + \dot{m}_{w1}=\dot{m}_{a2}w_2 + \dot{m}_{w2}
\end{equation}

siendo $w_1$ y $w_2$ la humedad absoluta a la entrada y la salida, $\dot{m}_{a1}$ y $\dot{m}_{a2}$ los flujos másicos de aire seco a la entrada y la salida y $\dot{m}_{w1}$ y $\dot{m}_{w2}$ los de agua.

### *Ejemplo de cálculos y muestra de resultados*

```python
p = 10**5 # Pa
T = 25 + 273.15 # K
R = 8.314 # J/(mol K)
M = 0.029 # kg/mol
V = 1 # m^3

# Cálculo del flujo másico de aire

m = p * V * M / (R * T)

print(f"El flujo másico de aire es {m:.2f} kg/s")
```

El flujo másico de aire es $1.17$ $\mathrm{kg/s}$ (unidades separadas del valor y en caracteres rectos)
