```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data for the plot
w1 = np.linspace(-3, -0.25, 200)
w2 = np.linspace(-0.25, 0.25, 200)
w3 = np.linspace(0.25, 3, 200)

# Define the loss function for the outer parts (black parts)
L_w1 = (w1**4 - 4*w1**2 + 6)
L_w3 = (w3**4 - 4*w3**2 + 6)

# Define a different function for the middle part (black part)
L_w2 = 892.0625*w2**2 - 50

# Generate random green points within the same ranges
np.random.seed(6)  # Set seed for reproducibility
green_points_w = np.random.uniform(-3, 3, 6)

# Calculate the corresponding loss values for the green points
green_points_L_w = []
for gp in green_points_w:
    if gp < -0.25:
        L = L_w1[np.argmin(abs(w1 - gp))]
    elif gp > 0.25:
        L = L_w3[np.argmin(abs(w3 - gp))]
    else:
        L = L_w2[np.argmin(abs(w2 - gp))]
    green_points_L_w.append(L)

# Plot the curve
plt.figure(figsize=(8, 6))
plt.plot(w1, L_w1, 'k-')  # Black part of the curve
plt.plot(w2, L_w2, 'k-')  # Black part of the curve
plt.plot(w3, L_w3, 'k-')  # Black part of the curve

# Plot the green points
plt.scatter(green_points_w, green_points_L_w, color='green')

plt.xlabel('$w$', fontsize=14)
plt.ylabel('$\mathcal{L}(w)$', fontsize=14)
plt.title('Loss Function Curve', fontsize=16)

# Hide the axis values
plt.xticks([])
plt.yticks([])

# Show the plot
plt.show()

```
